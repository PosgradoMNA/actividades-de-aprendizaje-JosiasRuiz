```python
'''
The two arguments for this function are the files:
    - currentMem: File containing list of current members
    - exMem: File containing list of old members
    
    This function should remove all rows from currentMem containing 'no' 
    in the 'Active' column and appends them to exMem.
    '''
def cleanFiles(currentMem, exMem):  
    with open(currentMem,'r+') as readfile:#TODO: Open the exMem file in r+ mode
        members = readfile.readlines() #TODO: Read each member in the currentMem (1 member per row) file into a list.
    
    with open(exMem,'a+') as appendfile:#TODO: Open the exMem file in a+ mode        
        with open(currentMem,'w+') as writefile: #TODO: Open the exMem file in w+ mode
            header=members[0]
            members.pop(0)
            # Hint: Recall that the first line in the file is the header.
            
            inactive=[]  #TODO: iterate through the members and create a new list of the innactive members
            for member in members:
                if 'no' in member:
                    inactive.append(member)
           
            writefile.seek(0) # Go to the beginning of the currentMem file
            writefile.write(header)
            
            for member in members:  # TODO: Iterate through the members list. 
                if (member in inactive):   # If a member is inactive, add them to exMem, otherwise write them into currentMem
                    appendfile.write(member)
                else:
                    writefile.write(member)            
    
   #pass # Remove this line when done implementation


# The code below is to help you view the files.
# Do not modify this code for this exercise.
memReg = 'members.txt'
exReg = 'inactive.txt'
cleanFiles(memReg,exReg)


headers = "Membership No  Date Joined  Active  \n"
with open(memReg,'r') as readFile:
    print("Active Members: \n\n")
    print(readFile.read())
    
with open(exReg,'r') as readFile:
    print("Inactive Members: \n\n")
    print(readFile.read())
```
